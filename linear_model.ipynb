```python

DF['Research'] = DF['Research'].apply(lambda x: str(x))

fig = px.scatter(DF,
                 x='CGPA',
                 y='Chance of Admit ',
                trendline='ols',
                color = 'Research',
                 
                )


fig.update_layout(template="ggplot2",
                  title='CGPA vs Chance of Admit Colored by Research Experience',
                  title_x=0.5,
                  xaxis_title="CGPA",
                  yaxis_title="Chance of Admission")

fig.show()
```

```python
fig = px.scatter(DF,
                 x='GRE Score',
                 y='Chance of Admit ',
                trendline='ols',
                 
                )


fig.update_layout(template="ggplot2",
                  title='GRE Score vs Chance of Admit',
                  title_x=0.5,
                  xaxis_title="GRE",
                  yaxis_title="Chance of Admission")

fig.show()
```

```python
my_columns = ['GRE Score', 'Chance of Admit ', 'CGPA']
DF_model=DF[my_columns]
DF_model
```

```python
X = DF_model['GRE Score'].values.reshape(-1,1)
y = DF_model['Chance of Admit ']

LM = LinearRegression()
LM.fit(X,y)

print(LM.coef_)
print(LM.intercept_)
print(LM.score(X,y))

# Add the line our model would predict
DF_model['predicted'] = LM.coef_*DF_model['GRE Score']+LM.intercept_
DF_model = DF_model.sort_values('predicted')
```

```python
fig = px.scatter(DF_model,
                 x='GRE Score',
                 y='Chance of Admit ',                 
                )


fig.add_trace(
    px.line(DF_model, x='GRE Score', y = 'predicted', color_discrete_sequence=['red']).data[0]
)

fig.show()
```

```python
X = DF_model['CGPA'].values.reshape(-1,1)
y = DF_model['Chance of Admit ']

LM = LinearRegression()
LM.fit(X,y)

print(LM.coef_)
print(LM.intercept_)
print(LM.score(X,y))

# Add the line our model would predict
DF_model['predicted'] = LM.coef_*DF_model['CGPA']+LM.intercept_
DF_model = DF_model.sort_values('predicted')
```

```python
fig = px.scatter(DF_model,
                 x='CGPA',
                 y='Chance of Admit ',                 
                )


fig.add_trace(
    px.line(DF_model, x='CGPA', y = 'predicted', color_discrete_sequence=['red']).data[0]
)

fig.show()
```
